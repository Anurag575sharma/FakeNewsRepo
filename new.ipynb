```python
# ──────────────────────────────────────────────────────────────────────────────
# 0) INSTALL DEPENDENCIES (run once)
# ──────────────────────────────────────────────────────────────────────────────
!pip install tensorflow pandas scikit-learn matplotlib scikit-fuzzy

# ──────────────────────────────────────────────────────────────────────────────
# 1) IMPORTS
# ──────────────────────────────────────────────────────────────────────────────
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Embedding, Bidirectional, LSTM, Dropout, Dense
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.preprocessing   import StandardScaler
from sklearn.metrics         import confusion_matrix, roc_curve, auc, precision_recall_curve, classification_report, accuracy_score
from sklearn.utils           import class_weight
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.manifold        import TSNE
import skfuzzy as fuzz
from scipy.stats             import mode

# ──────────────────────────────────────────────────────────────────────────────
# 2) LOAD & PREPROCESS LIAR DATASET
# ──────────────────────────────────────────────────────────────────────────────
# 2.1 Define column names
cols = [
    'id','label','statement','subject','speaker','job_title',
    'state','party','barely_true_counts','false_counts','half_true_counts',
    'mostly_true_counts','pants_on_fire_counts','context'
]

# 2.2 Read the CSVs (tab-separated, no header row)
train_df = pd.read_csv("train.csv", sep='\t', names=cols, header=None)
val_df   = pd.read_csv("valid.csv", sep='\t', names=cols, header=None)
test_df  = pd.read_csv("test.csv",  sep='\t', names=cols, header=None)

# 2.3 Binary-ify the multiclass label
def simplify_label(x):
    try:
        return 1 if x.lower() in ('true','mostly-true','half-true') else 0
    except:
        return 0

for df in (train_df, val_df, test_df):
    df['label']     = df['label'].apply(simplify_label)
    df['statement'] = df['statement'].fillna('').astype(str)
    df.drop(columns=[c for c in df.columns if c not in ('label','statement')], inplace=True)

# 2.4 Quick class balance check
print("Train class counts:\n", train_df['label'].value_counts(), "\n")

# ──────────────────────────────────────────────────────────────────────────────
# 3) TOKENIZE & PAD SEQUENCES
# ──────────────────────────────────────────────────────────────────────────────
all_text = pd.concat([train_df.statement, val_df.statement, test_df.statement])
tokenizer = Tokenizer(num_words=10000, oov_token="<OOV>")
tokenizer.fit_on_texts(all_text)

maxlen = 50
def make_padded(df):
    seq = tokenizer.texts_to_sequences(df.statement)
    return pad_sequences(seq, maxlen=maxlen, padding='post', truncating='post')

X_train, X_val, X_test = make_padded(train_df), make_padded(val_df), make_padded(test_df)
y_train, y_val, y_test = train_df.label.values, val_df.label.values, test_df.label.values

# ──────────────────────────────────────────────────────────────────────────────
# 4) EMBEDDING SETUP (GloVe if present, else trainable embeddings)
# ──────────────────────────────────────────────────────────────────────────────
vocab_size    = min(10000, len(tokenizer.word_index)+1)
embedding_dim = 100

if os.path.exists("glove.6B.100d.txt"):
    print("Loading GloVe embeddings...")
    emb_index = {}
    with open("glove.6B.100d.txt", encoding="utf8") as f:
        for line in f:
            vals = line.split()
            emb_index[vals[0]] = np.asarray(vals[1:], dtype='float32')
    emb_matrix = np.zeros((vocab_size, embedding_dim))
    for w,i in tokenizer.word_index.items():
        if i < vocab_size and w in emb_index:
            emb_matrix[i] = emb_index[w]
    embedding_layer = Embedding(
        vocab_size, embedding_dim,
        weights=[emb_matrix],
        input_length=maxlen,
        trainable=False
    )
else:
    print("GloVe not found, using trainable embeddings.")
    embedding_layer = Embedding(vocab_size, embedding_dim, input_length=maxlen, trainable=True)

# ──────────────────────────────────────────────────────────────────────────────
# 5) BUILD Bi-LSTM MODEL FUNCTION
# ──────────────────────────────────────────────────────────────────────────────
def build_model():
    m = Sequential([
        embedding_layer,
        Bidirectional(LSTM(64, return_sequences=False)),
        Dropout(0.5),
        Dense(64, activation='relu'),
        Dropout(0.3),
        Dense(1, activation='sigmoid')
    ])
    m.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return m

# 5.1 Instantiate and summarize
model = build_model()
model.summary()

# ──────────────────────────────────────────────────────────────────────────────
# 6) TRAIN WITH CALLBACKS & CLASS WEIGHTS
# ──────────────────────────────────────────────────────────────────────────────
# 6.1 Compute class weights
cw_vals = class_weight.compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
class_weights = dict(enumerate(cw_vals))

# 6.2 Callbacks
es  = EarlyStopping(monitor='val_accuracy', patience=3, restore_best_weights=True)
rlp = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2)

# 6.3 Train
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=10, batch_size=64,
    class_weight=class_weights,
    callbacks=[es, rlp]
)

# ──────────────────────────────────────────────────────────────────────────────
# 7) BASELINE EVALUATION
# ──────────────────────────────────────────────────────────────────────────────
probs       = model.predict(X_test).ravel()
y_pred_base = (probs >= 0.5).astype(int)

print("\n-- BiLSTM Baseline Classification Report --")
print(classification_report(y_test, y_pred_base))
base_acc = accuracy_score(y_test, y_pred_base)

# ──────────────────────────────────────────────────────────────────────────────
# 8) FCM REFINEMENT
# ──────────────────────────────────────────────────────────────────────────────
# 8.1 Extract 64-dim feature vectors
feat_extractor = Model(inputs=model.input, outputs=model.layers[-3].output)
X_feats        = feat_extractor.predict(X_test)

# 8.2 Normalize & transpose
X_norm = StandardScaler().fit_transform(X_feats).T

# 8.3 Fuzzy C-Means
cntr, u, *_ = fuzz.cluster.cmeans(X_norm, c=2, m=2, error=0.005, maxiter=1000)
fcm_labels  = np.argmax(u, axis=0)

# 8.4 Align cluster IDs by average Bi-LSTM prob
cluster_means = [np.mean(probs[fcm_labels==c]) if np.any(fcm_labels==c) else 0 for c in (0,1)]
real_cluster  = int(np.argmax(cluster_means))
y_pred_fcm    = (fcm_labels == real_cluster).astype(int)

print("\n-- FCM-Refined Classification Report --")
print(classification_report(y_test, y_pred_fcm))
refined_acc = accuracy_score(y_test, y_pred_fcm)

# # ──────────────────────────────────────────────────────────────────────────────
# # 9) STRATIFIED 3-FOLD CROSS-VALIDATION
# # ──────────────────────────────────────────────────────────────────────────────
# # Wrap for sklearn
# try:
#     from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
# except:
#     from keras.wrappers.scikit_learn import KerasClassifier

# clf = KerasClassifier(build_fn=build_model, epochs=5, batch_size=64, verbose=0)
# cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
# cv_scores = cross_val_score(clf, X_train, y_train, cv=cv,
#                             fit_params={'class_weight': class_weights})
# print(f"\n3-Fold CV Accuracies: {np.round(cv_scores,4)}")
# print(f"Mean CV Acc: {cv_scores.mean():.4f}, Std: {cv_scores.std():.4f}")

# ──────────────────────────────────────────────────────────────────────────────
# 9) STRATIFIED 3-FOLD CROSS-VALIDATION (with SciKeras)
# ──────────────────────────────────────────────────────────────────────────────
from scikeras.wrappers import KerasClassifier

# Define the KerasClassifier with updated parameters
clf = KerasClassifier(
    model=build_model,
    epochs=5,
    batch_size=64,
    verbose=0,
    class_weight=class_weights  # can be passed directly in SciKeras
)

# Stratified K-Fold
cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

# Evaluate
cv_scores = cross_val_score(clf, X_train, y_train, cv=cv)
print(f"\n3-Fold CV Accuracies: {np.round(cv_scores,4)}")
print(f"Mean CV Acc: {cv_scores.mean():.4f}, Std: {cv_scores.std():.4f}")


# ──────────────────────────────────────────────────────────────────────────────
# 10) t-SNE VISUALIZATION OF LSTM FEATURES
# ──────────────────────────────────────────────────────────────────────────────
X_2d = TSNE(n_components=2, random_state=42).fit_transform(X_feats)

# ──────────────────────────────────────────────────────────────────────────────
# 11) PLOTTING (7 Figures)
# ──────────────────────────────────────────────────────────────────────────────
# 11.1 Label distribution
plt.figure(figsize=(6,4))
counts = [train_df.label.value_counts().sort_index(),
          val_df.label.value_counts().sort_index(),
          test_df.label.value_counts().sort_index()]
x = np.arange(2)
w = 0.25
labels = ['Train','Val','Test']
for i,cnt in enumerate(counts):
    plt.bar(x + (i-1)*w, cnt, w, label=labels[i])
plt.xticks(x, ['Real (0)','Fake (1)'])
plt.ylabel('Count')
plt.title('Class Distribution')
plt.legend(); plt.tight_layout(); plt.show()

# 11.2 Accuracy over epochs
plt.figure(figsize=(6,4))
plt.plot(history.history['accuracy'],    marker='o', label='Train Acc')
plt.plot(history.history['val_accuracy'],marker='s', label='Val Acc')
plt.title('Accuracy over Epochs')
plt.xlabel('Epoch'); plt.ylabel('Accuracy')
plt.legend(); plt.grid(True); plt.tight_layout(); plt.show()

# 11.3 Loss over epochs
plt.figure(figsize=(6,4))
plt.plot(history.history['loss'],    marker='o', label='Train Loss')
plt.plot(history.history['val_loss'],marker='s', label='Val Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epoch'); plt.ylabel('Loss')
plt.legend(); plt.grid(True); plt.tight_layout(); plt.show()

# 11.4 Confusion matrix — baseline
cm_b = confusion_matrix(y_test, y_pred_base)
plt.figure(figsize=(4,4))
plt.imshow(cm_b, cmap='Blues')
plt.title('Confusion Matrix (Baseline)')
plt.xticks([0,1], ['Real','Fake']); plt.yticks([0,1], ['Real','Fake'])
for i in range(2):
    for j in range(2):
        plt.text(j,i,cm_b[i,j],ha='center',va='center',
                 color='white' if cm_b[i,j]>cm_b.max()/2 else 'black')
plt.tight_layout(); plt.show()

# 11.5 Confusion matrix — FCM refined
cm_r = confusion_matrix(y_test, y_pred_fcm)
plt.figure(figsize=(4,4))
plt.imshow(cm_r, cmap='Blues')
plt.title('Confusion Matrix (FCM Refined)')
plt.xticks([0,1], ['Real','Fake']); plt.yticks([0,1], ['Real','Fake'])
for i in range(2):
    for j in range(2):
        plt.text(j,i,cm_r[i,j],ha='center',va='center',
                 color='white' if cm_r[i,j]>cm_r.max()/2 else 'black')
plt.tight_layout(); plt.show()

# 11.6 ROC curves
fpr_b, tpr_b, _ = roc_curve(y_test, probs)
fpr_r, tpr_r, _ = roc_curve(y_test, u[real_cluster])
auc_b = auc(fpr_b, tpr_b)
auc_r = auc(fpr_r, tpr_r)
plt.figure(figsize=(6,4))
plt.plot(fpr_b, tpr_b, label=f'Baseline (AUC={auc_b:.2f})')
plt.plot(fpr_r, tpr_r, label=f'FCM Refined (AUC={auc_r:.2f})')
plt.plot([0,1],[0,1],'--',color='gray')
plt.title('ROC Curve')
plt.xlabel('FPR'); plt.ylabel('TPR')
plt.legend(); plt.grid(True); plt.tight_layout(); plt.show()

# 11.7 Precision-Recall curves
prec_b, rec_b, _ = precision_recall_curve(y_test, probs)
prec_r, rec_r, _ = precision_recall_curve(y_test, u[real_cluster])
pr_auc_b = auc(rec_b, prec_b)
pr_auc_r = auc(rec_r, prec_r)
plt.figure(figsize=(6,4))
plt.plot(rec_b, prec_b, label=f'Baseline (AUC={pr_auc_b:.2f})')
plt.plot(rec_r, rec_r, label=f'FCM Refined (AUC={pr_auc_r:.2f})')
plt.title('Precision-Recall Curve')
plt.xlabel('Recall'); plt.ylabel('Precision')
plt.legend(); plt.grid(True); plt.tight_layout(); plt.show()
```

```python
# ──────────────────────────────────────────────────────────────────────────────
# 0) INSTALL DEPENDENCIES (run once)
# ──────────────────────────────────────────────────────────────────────────────
!pip install tensorflow pandas scikit-learn matplotlib scikit-fuzzy scikeras

# ──────────────────────────────────────────────────────────────────────────────
# 1) IMPORTS
# ──────────────────────────────────────────────────────────────────────────────
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Embedding, Bidirectional, LSTM, Dropout, Dense
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.preprocessing   import StandardScaler
from sklearn.metrics         import confusion_matrix, roc_curve, auc, precision_recall_curve, classification_report, accuracy_score
from sklearn.utils           import class_weight
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.manifold        import TSNE
import skfuzzy as fuzz
from scipy.stats             import mode
from scikeras.wrappers import KerasClassifier

# ──────────────────────────────────────────────────────────────────────────────
# 2) LOAD & PREPROCESS LIAR DATASET
# ──────────────────────────────────────────────────────────────────────────────
# 2.1 Define column names
cols = [
    'id','label','statement','subject','speaker','job_title',
    'state','party','barely_true_counts','false_counts','half_true_counts',
    'mostly_true_counts','pants_on_fire_counts','context'
]

# 2.2 Read the CSVs (tab-separated, no header row)
train_df = pd.read_csv("train.csv", sep='\t', names=cols, header=None)
val_df   = pd.read_csv("valid.csv", sep='\t', names=cols, header=None)
test_df  = pd.read_csv("test.csv",  sep='\t', names=cols, header=None)

# 2.3 Binary-ify the multiclass label
def simplify_label(x):
    if pd.isnull(x):  # Check if the value is NaN
        return 0
    return 1 if x.lower() in ('true', 'mostly-true', 'half-true') else 0

for df in (train_df, val_df, test_df):
    df['label']     = df['label'].apply(simplify_label)
    df['statement'] = df['statement'].fillna('').astype(str)
    df.drop(columns=[c for c in df.columns if c not in ('label','statement')], inplace=True)

# 2.4 Quick class balance check
print("Train class counts:\n", train_df['label'].value_counts(), "\n")

# ──────────────────────────────────────────────────────────────────────────────
# 3) TOKENIZE & PAD SEQUENCES
# ──────────────────────────────────────────────────────────────────────────────
all_text = pd.concat([train_df.statement, val_df.statement, test_df.statement])
tokenizer = Tokenizer(num_words=10000, oov_token="<OOV>")
tokenizer.fit_on_texts(all_text)

maxlen = 50
def make_padded(df):
    seq = tokenizer.texts_to_sequences(df.statement)
    return pad_sequences(seq, maxlen=maxlen, padding='post', truncating='post')

X_train, X_val, X_test = make_padded(train_df), make_padded(val_df), make_padded(test_df)
y_train, y_val, y_test = train_df.label.values, val_df.label.values, test_df.label.values

# ──────────────────────────────────────────────────────────────────────────────
# 4) EMBEDDING SETUP (GloVe if present, else trainable embeddings)
# ──────────────────────────────────────────────────────────────────────────────
vocab_size    = min(10000, len(tokenizer.word_index)+1)
embedding_dim = 100

if os.path.exists("glove.6B.100d.txt"):
    print("Loading GloVe embeddings...")
    emb_index = {}
    with open("glove.6B.100d.txt", encoding="utf8") as f:
        for line in f:
            vals = line.split()
            emb_index[vals[0]] = np.asarray(vals[1:], dtype='float32')
    emb_matrix = np.zeros((vocab_size, embedding_dim))
    for w,i in tokenizer.word_index.items():
        if i < vocab_size and w in emb_index:
            emb_matrix[i] = emb_index[w]
    embedding_layer = Embedding(
        vocab_size, embedding_dim,
        weights=[emb_matrix],
        input_length=maxlen,
        trainable=False
    )
else:
    print("GloVe not found, using trainable embeddings.")
    embedding_layer = Embedding(vocab_size, embedding_dim, input_length=maxlen, trainable=True)

# ──────────────────────────────────────────────────────────────────────────────
# 5) BUILD Bi-LSTM MODEL FUNCTION
# ──────────────────────────────────────────────────────────────────────────────
def build_model():
    m = Sequential([
        embedding_layer,
        Bidirectional(LSTM(64, return_sequences=False)),
        Dropout(0.5),
        Dense(64, activation='relu'),
        Dropout(0.3),
        Dense(1, activation='sigmoid')
    ])
    m.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return m

# 5.1 Instantiate and summarize
model = build_model()
model.summary()

# ──────────────────────────────────────────────────────────────────────────────
# 6) TRAIN WITH CALLBACKS & CLASS WEIGHTS
# ──────────────────────────────────────────────────────────────────────────────
# 6.1 Compute class weights
cw_vals = class_weight.compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
class_weights = dict(enumerate(cw_vals))

# 6.2 Callbacks
es  = EarlyStopping(monitor='val_accuracy', patience=3, restore_best_weights=True)
rlp = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2)

# 6.3 Train
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=10, batch_size=64,
    class_weight=class_weights,
    callbacks=[es, rlp]
)

# ──────────────────────────────────────────────────────────────────────────────
# 7) BASELINE EVALUATION
# ──────────────────────────────────────────────────────────────────────────────
probs       = model.predict(X_test).ravel()
y_pred_base = (probs >= 0.5).astype(int)

print("\n-- BiLSTM Baseline Classification Report --")
print(classification_report(y_test, y_pred_base))
base_acc = accuracy_score(y_test, y_pred_base)

# ──────────────────────────────────────────────────────────────────────────────
# 8) FCM REFINEMENT
# ──────────────────────────────────────────────────────────────────────────────
# 8.1 Extract 64-dim feature vectors
feat_extractor = Model(inputs=model.input, outputs=model.layers[-3].output)
X_feats        = feat_extractor.predict(X_test)

# 8.2 Normalize & transpose
X_norm = StandardScaler().fit_transform(X_feats).T

# 8.3 Fuzzy C-Means
cntr, u, *_ = fuzz.cluster.cmeans(X_norm, c=2, m=2, error=0.005, maxiter=1000)
fcm_labels  = np.argmax(u, axis=0)

# 8.4 Align cluster IDs by average Bi-LSTM prob
cluster_means = [np.mean(probs[fcm_labels==c]) if np.any(fcm_labels==c) else 0 for c in (0,1)]
real_cluster  = int(np.argmax(cluster_means))
y_pred_fcm    = (fcm_labels == real_cluster).astype(int)

print("\n-- FCM-Refined Classification Report --")
print(classification_report(y_test, y_pred_fcm))
refined_acc = accuracy_score(y_test, y_pred_fcm)

# # ──────────────────────────────────────────────────────────────────────────────
# # 9) STRATIFIED 3-FOLD CROSS-VALIDATION (with SciKeras)
# # ──────────────────────────────────────────────────────────────────────────────
# clf = KerasClassifier(
#     model=build_model,
#     epochs=5,
#     batch_size=64,
#     verbose=0,
#     class_weight=class_weights  # can be passed directly in SciKeras
# )

# # Stratified K-Fold
# cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

# # Evaluate
# cv_scores = cross_val_score(clf, X_train, y_train, cv=cv)
# print(f"\n3-Fold CV Accuracies: {np.round(cv_scores,4)}")
# print(f"Mean CV Acc: {cv_scores.mean():.4f}, Std: {cv_scores.std():.4f}")

# ──────────────────────────────────────────────────────────────────────────────
# 10) t-SNE VISUALIZATION OF LSTM FEATURES
# ──────────────────────────────────────────────────────────────────────────────
X_2d = TSNE(n_components=2, random_state=42).fit_transform(X_feats)

# ──────────────────────────────────────────────────────────────────────────────
# 11) PLOTTING (7 Figures)
# ──────────────────────────────────────────────────────────────────────────────
# 11.1 Label distribution
plt.figure(figsize=(6,4))
counts = [train_df.label.value_counts().sort_index(),
          val_df.label.value_counts().sort_index(),
          test_df.label.value_counts().sort_index()]
x = np.arange(2)
w = 0.25
labels = ['Train','Val','Test']
for i,cnt in enumerate(counts):
    plt.bar(x + (i-1)*w, cnt, w, label=labels[i])
plt.xticks(x, ['Real (0)','Fake (1)'])
plt.ylabel('Count')
plt.title('Class Distribution')
plt.legend(); plt.tight_layout(); plt.show()

# 11.2 Accuracy over epochs
plt.figure(figsize=(6,4))
plt.plot(history.history['accuracy'],    marker='o', label='Train Acc')
plt.plot(history.history['val_accuracy'],marker='s', label='Val Acc')
plt.title('Accuracy over Epochs')
plt.xlabel('Epoch'); plt.ylabel('Accuracy')
plt.legend(); plt.grid(True); plt.tight_layout(); plt.show()

# 11.3 Loss over epochs
plt.figure(figsize=(6,4))
plt.plot(history.history['loss'],    marker='o', label='Train Loss')
plt.plot(history.history['val_loss'],marker='s', label='Val Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epoch'); plt.ylabel('Loss')
plt.legend(); plt.grid(True); plt.tight_layout(); plt.show()

# 11.4 Confusion matrix — baseline
cm_b = confusion_matrix(y_test, y_pred_base)
plt.figure(figsize=(4,4))
plt.imshow(cm_b, cmap='Blues')
plt.title('Confusion Matrix (Baseline)')
plt.xticks([0,1], ['Real','Fake']); plt.yticks([0,1], ['Real','Fake'])
for i in range(2):
    for j in range(2):
        plt.text(j,i,cm_b[i,j],ha='center',va='center',
                 color='white' if cm_b[i,j]>cm_b.max()/2 else 'black')
plt.tight_layout(); plt.show()

# 11.5 Confusion matrix — FCM refined
cm_r = confusion_matrix(y_test, y_pred_fcm)
plt.figure(figsize=(4,4))
plt.imshow(cm_r, cmap='Blues')
plt.title('Confusion Matrix (FCM Refined)')
plt.xticks([0,1], ['Real','Fake']); plt.yticks([0,1], ['Real','Fake'])
for i in range(2):
    for j in range(2):
        plt.text(j,i,cm_r[i,j],ha='center',va='center',
                 color='white' if cm_r[i,j]>cm_r.max()/2 else 'black')
plt.tight_layout(); plt.show()

# 11.6 ROC curve — baseline
fpr, tpr, _ = roc_curve(y_test, probs)
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(6,4))
plt.plot(fpr, tpr, label=f'AUC = {roc_auc:.4f}')
plt.plot([0,1], [0,1], 'k--')
plt.title('ROC Curve (Baseline)')
plt.xlabel('False Positive Rate'); plt.ylabel('True Positive Rate')
plt.legend(); plt.grid(True); plt.tight_layout(); plt.show()

# 11.7 ROC curve — FCM refined
fpr_r, tpr_r, _ = roc_curve(y_test, probs)
roc_auc_r = auc(fpr_r, tpr_r)
plt.figure(figsize=(6,4))
plt.plot(fpr_r, tpr_r, label=f'AUC = {roc_auc_r:.4f}')
plt.plot([0,1], [0,1], 'k--')
plt.title('ROC Curve (FCM Refined)')
plt.xlabel('False Positive Rate'); plt.ylabel('True Positive Rate')
plt.legend(); plt.grid(True); plt.tight_layout(); plt.show()

```
